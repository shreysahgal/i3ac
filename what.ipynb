```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('embeddings.npy', allow_pickle=True)
```

```python
for i in range(0,data.shape[0], 10):
    # save data[i][state][0] and data[i][new_state][0] to a file as images side by side
    fig, ax = plt.subplots(1, 2)
    ax[0].imshow(data[i]['state'][0])
    ax[1].imshow(data[i]['next_state'][0])
    ax[0].axis('off')
    ax[1].axis('off')
    ax[0].set_title('State')
    ax[1].set_title('New State')
    ax[0].set_aspect('equal')
    fig.savefig('images/{}.png'.format(i))
```
